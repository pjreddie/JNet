```python
%matplotlib inline

import cv2
from tiler import *
from matplotlib import pyplot as plt
```

```python
img = read_image('images/dog.jpg')
```

```python
boxes = image_boxes(img, [416,416])
```

```python
len(boxes)
```

```python
#show_image(boxes[0]['img'], wait=False)
```

```python
im = cv2.cvtColor(boxes[0]['img'], cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(im)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
for i in range(len(boxes)):
    im = cv2.cvtColor(boxes[i]['img'], cv2.COLOR_BGR2RGB)
    if im.shape[0]==416 and im.shape[1]==416:
        cv2.imwrite('temp.jpg', cv2.cvtColor(im, cv2.COLOR_RGB2BGR))
        #plt.figure(figsize=(5,5))
        plt.imshow(im)
        plt.xticks([])
        plt.yticks([])
        plt.savefig('temp.png')
        plt.title('res: '+str(im.shape)+'\n'+'pos: '+str(boxes[i]['pos']))
        #plt.show()
        break
```
